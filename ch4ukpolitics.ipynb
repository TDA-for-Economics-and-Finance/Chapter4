# UK Politics Data

This notebook works through the examples in Chapter 4 which draw upon the UK politics data. More information about the data is within the Dlotko and Rudkin (2024) book. 

To run this code the file ukpolitics.csv is needed

As with all examples we begin by importing the necessary packages

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

Next we import the data

```python
data = pd.read_csv('ukpolitics.csv')
```

Let us take a look at the data with the head() function. We confirm that all variables are measured in percentages.

```python
data.head()
```

We create two dummies. Firstly Con19 which takes the value 1 if the clab19 variable is positive. Secondly, LC, which takes the value 1 if the Leave vote (leave) is greater than 50.

The () contain the if condition and the *1 ensures that the result returned is a numeric value

```python
data['Con19'] = (data['clab19']>0)*1
data['LC'] = (data['leave']>50)*1
```

Let us first create a scatter plot in which we show the Leave percentages and Conservative lead in 2019 on the axes

```python
fig, ax = plt.subplots(figsize = ( 5 , 5 )) 
  
# Plot the scatterplot 
sb.scatterplot( ax = ax , x = 'leave' , y = 'clab19' , data = data ) 
  
# Set label for x-axis 
ax.set_xlabel( "Leave Percentage" , size = 12 ) 
  
# Set label for y-axis 
ax.set_ylabel( "Conservative - Labour 2019" , size = 12 ) 
  
# Display figure 
plt.show() 
```

To check our dummies let us color this scatter plot according to the values of LC and Con19

```python
fig, ax = plt.subplots(figsize = ( 5 , 5 )) 
  
# Plot the scatterplot 
sb.scatterplot( ax = ax , x = 'leave' , y = 'clab19' , data = data , hue= 'LC') 
  
# Set label for x-axis 
ax.set_xlabel( "Leave Percentage" , size = 12 ) 
  
# Set label for y-axis 
ax.set_ylabel( "Conservative - Labour 2019" , size = 12 ) 
  
# Display figure 
plt.show() 
```

```python
fig, ax = plt.subplots(figsize = ( 5 , 5 )) 
  
# Plot the scatterplot 
sb.scatterplot( ax = ax , x = 'leave' , y = 'clab19' , data = data , hue= 'Con19') 
  
# Set label for x-axis 
ax.set_xlabel( "Leave Percentage" , size = 12 ) 
  
# Set label for y-axis 
ax.set_ylabel( "Conservative - Labour 2019" , size = 12 ) 
  
# Display figure 
plt.show() 
```

# Distributions

Here we apply the univariate and bivariate distribution plots from Seaborn to examine the relationships within the data. The dummy variables LC and Con19 are used to illustrate how coloration can be added to distribution plots when a suitable coloration variable exists.

The first two examples look at the distribution of the Leave vote and color according to whether the Conservatives achieved a higher vote than Labour in the 2019 election

```python
sb.displot(data, x='leave',hue="Con19")
```

```python
sb.displot(data, x='leave',hue="Con19", kind="kde")
```

Let us now return to the question of whether owning a home with a mortgage makes a difference to the voting behaviour. The distributions of mortage provide the horizontal axis and Conservative voting the vertical axis. The coloration is according to whether the consituency was estimated to vote Leave by Hanretty (2017)

You are welcome to play with this code to try other variables

```python
sb.jointplot(
    data=data,
    x="mortgage", y="clab19", hue="LC"
)
```

We may enhance the joint distribution plot with box plots for the individual distributions. Here we use the Conservative lead in 2019 and the Hanretty (2017) estimated leave percentage. Boxplots are added to show the marginal distributions.

Again you are welcome to play with the code to add coloration, or change to kernel density (kde)

```python
g = sb.JointGrid(data=data, x="leave", y="clab19")
g.plot_joint(sb.histplot)
g.plot_marginals(sb.boxplot)
```

```python
g = sb.catplot(
    data=data.melt(id_vars = ["region"], value_vars=["clab19", "leave", "mortgage"]),
    kind="bar",
    y="value", x="variable", col="region", errorbar=None, col_wrap=4
)
g.set_axis_labels("", "")
```

The previous example shows how a bad bar chart can be created. The Leave percentages dominate because they are always above 30. It is more interesting to see how the region voted by comparing the leave vote with 50. Hence let us create a second leave variable, leave50, which makes that comparison.

We then redraw the plots

```python
data['leave50'] = data['leave']-50
```

```python
g = sb.catplot(
    data=data.melt(id_vars = ["region"], value_vars=["clab19", "leave50", "mortgage"]),
    kind="bar",
    y="value", x="variable", col="region", errorbar=None, col_wrap=4
)
g.set_axis_labels("", "")
```
